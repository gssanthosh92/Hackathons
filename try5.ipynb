```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
train_data  = pd.read_csv('Train.csv') #loads train data
test_data = pd.read_csv('Test.csv') #loads test data
```

```python
train_data.head()
#test_data.head()
```

```python
train_data=train_data.dropna()
train_data=train_data.drop(columns=['InvoiceDate','InvoiceNo'],axis=1)
train_data.head() #removing null values and unwanted columns
```

```python
test_data=test_data.dropna()
test_data=test_data.drop(columns=['InvoiceDate','InvoiceNo'],axis=1)
test_data.head() #removing null values and unwanted columns
```

```python
train_data.info()
new_data=train_data['UnitPrice']
new_data=new_data[0:122049]
new_data
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
regr = linear_model.LinearRegression()
x = np.asanyarray(train_data[['StockCode','Description','Quantity','CustomerID','Country']])
y = np.asanyarray(train_data[['UnitPrice']])
regr.fit (x, y)
final=regr.predict(test_data[['StockCode','Description','Quantity','CustomerID','Country']])
final #using multivariate regression to fit into a model and then predciting the values of test data
```

```python
result = pd.DataFrame(data=final).round(2)
result
result.columns=['UnitPrice']
result.set_index("UnitPrice", inplace = True)
result.to_csv("MachineHack_Prediction_5.csv") #converting the dataframe into CSV file with rounded values
```

```python
print("RMSE :",np.sqrt(mean_squared_error(new_data, final)))
```
